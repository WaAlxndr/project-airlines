```python
tmp_flight_df.drop_duplicates('tail_num')
```

 Tutaj nadpisz `flight_df` używając `tmp_flight_df`

```python
flight_df = tmp_flight_df
flight_df = flight_df.rename({'dep_delay_new' : 'dep_delay'}, axis='columns')
flight_df['is_delayed'] = flight_df['dep_delay'] > 15
flight_df
```

 ## Opóźnienia vs. rok produkcji samolotu cz. 1
 Tutaj wyznacz zależność roku produkcji do częstotliwości opóźnień

```python
flight_dep_delays_manufacture_year_by_count = flight_df[['dep_delay','manufacture_year']].groupby(by=['manufacture_year']).count()
flight_is_delays_manufacture_year_sum = flight_df[['is_delayed','manufacture_year']].groupby(by=['manufacture_year']).sum()

flight_delays_by_manufacture_year_df = pd.concat([flight_dep_delays_manufacture_year_by_count, flight_is_delays_manufacture_year_sum], axis=1)
flight_delays_by_manufacture_year_df['ratio_year'] = round(flight_delays_by_manufacture_year_df['is_delayed']/flight_delays_by_manufacture_year_df['dep_delay'],2)
flight_delays_by_manufacture_year_df
```

 Tutaj wyrysuj ramkę `delays_by_manufacture_year_df`

```python
delays_by_manufacture_year_df = flight_delays_by_manufacture_year_df.reset_index()
delays_by_manufacture_year_df
```

 Tutaj zmodyfikuj wykres tak, aby prezentował tylko te roczniki, które odbyły więcej niż 10 000 lotów

```python
delays_by_manufacture_year_df=delays_by_manufacture_year_df.loc[delays_by_manufacture_year_df['dep_delay'] > 10000]
```

```python
year=delays_by_manufacture_year_df['manufacture_year']
ax10 = delays_by_manufacture_year_df.plot(
    kind = 'bar',
    x= 'manufacture_year',
    y= 'ratio_year',
    figsize = (15,5),
    color = 'pink',
    legend = True)
ax10.set_xlabel('Manufacture year ', fontsize = 15)
ax10.set_ylabel('Delay ratio', fontsize = 15)
ax10.set_title('Manufacture year vs delay ratio', fontsize = 25)
ax10.set_xticklabels(year,rotation=45)
```
